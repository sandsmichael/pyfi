## Retreive

```python
from pyfi.lib.retrievers import fred, equity
# data = fred.get_matrix(ids = ['VIXCLS', *fred.IDS_STANDARD])
const = equity.get_sp500_constituents()[:20]
data = equity.get_price_matrix(tickers=const, start_date="2020-01-01", end_date="2024-01-01").dropna(how='any', axis=1)
```
